```python
#NOTE TO SELF: NEED TO GET LIST/TUPLE OF MOTIF COORDINATES AND REMOVE THEM

#this is only for DEL

def remove_substrings(seq, coordinates):
    result = ""
    index = 0
    
    for start, end in coordinates:
        result += seq[index:start]
        index = end + 1
    
    result += seq[index:]
    
    return result
```

```python
#THIS IS FOR DUP 

def duplicate_substrings(string, coordinates):
    result = ""
    index = 0
    
    for start, end in coordinates:
        result += string[index:start]  # Add substring before the duplicate
        substring_to_duplicate = string[start:end+1]  # Extract substring to duplicate
        result += substring_to_duplicate  # Add original substring
        result += substring_to_duplicate  # Add duplicated substring
        index = end + 1
    
    result += string[index:]  # Add remaining substring
    
    return result

# Example usage:
string = "abcdefgh"
coordinates_to_duplicate = [(1, 3), (5, 8)]  # Coordinates to duplicate (start, end)
result = duplicate_substrings(string, coordinates_to_duplicate)
print(result)  # 
```

```python
#THIS IS FOR INV 

def invert_substrings(string, coordinates):
    result = list(string)  # Convert the string to a list of characters for easier modification
    
    for start, end in coordinates:
        # Invert the substring within the list
        result[start:end+1] = result[start:end+1][::-1]
    
    # Convert the list back to a string
    return ''.join(result)

# Example usage:
string = "abcdefgh"
coordinates_to_invert = [(1, 3), (5, 8)]  # Coordinates to invert (start, end)
result = invert_substrings(string, coordinates_to_invert)
print(result)  # Output: 'acbdehgf'
```
