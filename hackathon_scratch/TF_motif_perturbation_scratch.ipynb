```python
#THIS IS FOR INV

def invert_motifs(sequence, substr_ranges):
    seq_list = list(sequence)  
    
    for start, end in substr_ranges:
        # Invert the substring within the list
        seq_list[start:end] = seq_list[start:end][::-1]
        
    # Convert the list back to a string
    return ''.join(seq_list)

# Example usage:
string = "abcdefgh"
coordinates_to_invert = [(1, 3), (5, 8)]  # Coordinates to invert (start, end)
result = invert_motifs(string, coordinates_to_invert)
print(result)  # Output: 'acbdehgf'
```

```python
#THIS IS FOR SHUFFLING
def shuffle_motifs(sequence, substr_ranges):
    seq_list = list(sequence)

    # Shuffle nucleotides within each substring
    for start, end in substr_ranges:
        # Extract the substring
        substring = seq_list[start:end]

        # Shuffle the substring in-place
        random.shuffle(substring)

        # Update the sequence with the shuffled substring
        seq_list[start:end] = substring
    
    return ''.join(seq_list)

# Example usage:
# this seems to work.... lol 
sequence = "ATGCATCGATCGATCG"
substr_ranges = [(2, 7), (11,16)]  # Example substring ranges
shuffled_sequence = shuffle_motifs(sequence, substr_ranges)
print(shuffled_sequence)
```

```python
#THIS IS FOR DUP 

def duplicate_substrings(string, coordinates):
    result = ""
    index = 0
    
    for start, end in coordinates:
        result += string[index:start]  # Add substring before the duplicate
        substring_to_duplicate = string[start:end+1]  # Extract substring to duplicate
        result += substring_to_duplicate  # Add original substring
        result += substring_to_duplicate  # Add duplicated substring
        index = end + 1
    
    result += string[index:]  # Add remaining substring
    
    return result

# Example usage:
string = "abcdefgh"
coordinates_to_duplicate = [(1, 3), (5, 8)]  # Coordinates to duplicate (start, end)
result = duplicate_substrings(string, coordinates_to_duplicate)
print(result)  # 
```
